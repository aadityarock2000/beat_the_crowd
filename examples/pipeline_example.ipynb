# Running the SmoothFly App



### 1) <ins>Python Installation</ins>: Python 3.10 is required in order to run this application. You can install this version by visiting the [Python download page](https://www.python.org/downloads/) and installing version 3.10 or above on either your windows or mac machine

### 2) <ins>Set Up</ins>: In order to collaborate on this project you should first clone the repo from https://github.com/aadityarock2000/beat_the_crowd. Once the repo has been cloned and you have navigated to the root directory, you can run the following commands to set the project.

### Windows:
### py -m pip install --upgrade build
### py -m build

### Unix/macOS:
### python3 -m pip install --upgrade build
### python3 -m build
